## 1. Introduction<a class="anchor" id="1"></a>

### 1.1 Objective <a class="anchor" id="1.1"></a>

The objective of this project is to classify mushrooms as edible or poisonous based on 22 descriptive attributes. The dataset contains 8124 observations. The descriptive features include measurements and observations for the cap, gills and stalk of the mushroom, as well as its population at the sample site, habitat and odor. Mushrooms were classed as edible or poisonous. Those that were regarded as 'not recommended' or 'edibility unknown' were classed as poisonous.<br><br>

The data was sourced from The Audubon Society Field Guide to North American Mushrooms (1981) and was made available on the UCI Machine learning database: https://archive.ics.uci.edu/ml/datasets/mushroom.

```python
target=mushroom.pop("class")
features=mushroom
features.head(5)
```

After tidying the column heading, the info() function confirms that all descriptive features are categorical - dtypes: objects(22)

```python
# tidy up the headings:
mushroom.columns = mushroom.columns.str.replace("[-]", "_")
mushroom.columns = mushroom.columns.str.replace("[%3F]", "")
mushroom.columns
# features information
features.info()
```

Frequency plots give an indication of possible problems or redundant data:

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
# frequency plots
columns = list(features)
for column in columns:
    count = features[column].value_counts()
    sns.set(style="darkgrid")
    sns.barplot(count.index, count.values, alpha=0.9)
    plt.title('Frequency Distribution of '+ column)
    plt.ylabel('Number of Occurrences', fontsize=12)
    plt.xlabel(column, fontsize=12)
    plt.show()
```

Veil_type has only one level: p (partial). It can therefore be removed from the dataset because it is not providing any information. For some features, it may be possible to collate levels with low frequency, for example; spore_print_color, without losing information.

## 3. Data Cleaning <a class="anchor" id="3"></a>

### 3.1 Redundant Values <a class="anchor" id="3.1"></a>

As mentioned above, veil_type can be removed because it does not provide any information.

```python
features=features.drop(columns="veil_type")
features.columns
```

### 3.2 Missing Values <a class="anchor" id="3.2"></a>

The only missing values in the dataset occur in the feature stalk_root.  There are 2480 missing values which represents 30.5% of that feature.  This feature may have high information gain, so rather than removing the feature completely, the missing values are replaced with 'u' for unknown.

```python
print(features.isnull().sum()) 
```

```python
features['stalk_root']= features['stalk_root'].fillna('u')
```

```python
count = features['stalk_root'].value_counts()
sns.set(style="darkgrid")
sns.barplot(count.index, count.values, alpha=0.9)
plt.title('Frequency Distribution of Stalk Root')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Stalk Root', fontsize=12)
plt.show()
```

A crosstab was created to see if any of the levels of stalk_root would split the data cleanly into one of the two target feature levels: poisonous or edible.

```python
cleaned_mushroom = features
cleaned_mushroom['class']=target
cleaned_mushroom.columns
pd.crosstab(cleaned_mushroom['stalk_root'],cleaned_mushroom['class'], margins=True)
# rooted = edible
```

Level 'r' in stalk_root refers to the characteristic described as 'rooted'. As this level splits the dataset cleanly into the edible class, stalk_root appears to offer some information gain despite the large number of missing variables.

To confirm that the dataset is now free of missing values:

```python
print(features.isnull().sum())
```

## 4. Further Analysis  <a class="anchor" id="4"></a>

Crosstabs were used to discover more about each descriptive feature and how it relates to the target feature.  Odor appears to be the feature that offers the most information about which mushrooms are poisonous as eight of its nine levels split the data into pure sets. The remaining level, 'n' indicating a lack of odor, is heavily skewed.

```python
pd.crosstab(cleaned_mushroom['odor'],cleaned_mushroom['class'], margins=True)
#odor very telling with every category except 'none' splitting the mushrooms into p and e cleanly
```

The features stalk_color_above_ring and stalk_color_below_ring split the dataset in almost the same way. In the two crosstabs, only levels n and w (brown and white) have different frequencies of poisonous and edible observations. However, the differences are slight and neither of these levels split the data into pure sets.  The other level with varying frequencies is that of 'y' or yellow.  In both crosstabs, a yellow color indicates a poisonous mushroom. However, the frequency in the crosstab for stalk_color_below_ring is 24 as opposed to 8 in the other.  As a result, the feature stalk_color_below_ring will be retained and the other removed.

```python
pd.crosstab(cleaned_mushroom['stalk_color_above_ring'],cleaned_mushroom['class'], margins=True)
# buff = p
# cinnamon = p
# red = e
# gray = e
# orange = e
# yellow = p
```

```python
pd.crosstab(cleaned_mushroom['stalk_color_below_ring'],cleaned_mushroom['class'], margins=True)
# buff = p
# cinnamon = p
# red = e
# gray = e
# orange = e
# yellow = p
```

Spore_print_color was mentioned earlier as a possible candidate for further processing because it had many levels with very low frequencies. The crosstab shows that these levels provide good information because they can split the data into pure sets.

```python
pd.crosstab(cleaned_mushroom['spore_print_color'],cleaned_mushroom['class'], margins=True)
# r b u y o have low frequencies but provide good info.
```

Many of the descriptive features have some levels that create pure sets and others that do not.  Gill_color is an example of this. Of the 12 levels, 4 create pure sets and the others created mixed sets. These levels may provide a high level of information, however, when combined with other features.

```python
pd.crosstab(cleaned_mushroom['gill_color'],cleaned_mushroom['class'], margins=True)
# buff = p
# green = p
# red = e
# orange = e
```

Creating a crosstab with two features that have large information gains (odor and stalk_color_below_ring) and the target feature creates a table that very neatly separates approximately 75% of observations into poisonous or edible classes.  The only observations that were not split into pure sets by these two descriptive features are brown below ring(n) with musty odor(m) and white below ring(w) with no odor(n). These observations total 2,064 and represent 25.4% of the total dataset.

```python
pd.crosstab([cleaned_mushroom['odor'],cleaned_mushroom['stalk_color_below_ring']], cleaned_mushroom['class'], margins=True)
```

The redundant feature: stalk_color_above_ring is removed.  21 descriptive features remain.

```python
features=features.drop(columns="stalk_color_above_ring")
```

## 6. Summary <a class="anchor" id="6"></a>

Missing values in the descriptive feature stalk_root were examined. A crosstab of stalk_root against the target feature revealed that a least one of the other levels in this feature provided a lot of information. All of the observations in the dataset had veil_type recorded as 'partial' so this feature was removed.<br><br>

All but two combinations of the levels of these two features were able to split the data into pure sets. The only level combination to not achieve this were mushrooms with brown stalk color below the ring and a musty odor and white stalk color below ring with no odor. This combination represents 25.4% of the dataset.<br><br>
